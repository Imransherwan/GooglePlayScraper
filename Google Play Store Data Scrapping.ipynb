### Function for all the above steps

```python
def retrieve_apps(category):
    
    result = search( category, lang="en", country="pk")
    
    # Create a list to store the data
    data = []
    for app in result:
        app_info = google_play_scraper.app(app["appId"])
        
        data.append({
            "Id":app["appId"],
              "title": app_info["title"],
              "summary": app_info["summary"],
              "number_of_installs": app_info["realInstalls"],
              "Average_rating": app_info["score"],
              "Ratings_recieved": app_info["ratings"],
              "Reviews_recieved": app_info["reviews"],
              "Ratings_histogram": app_info["histogram"],
              "price": app_info["price"],
              "developer_name": app_info["developer"],
              "genre": app_info["genre"]
        })
    
    df = pd.DataFrame(data)
    return df

```

```python
df = retrieve_apps("E-commerce")
df.to_csv("E-commerce_apps_info.csv",index=False)
```

```python
def scrap_reviews(input_csv_file, output_csv_file):
    # Read the input CSV file into a DataFrame
    df = pd.read_csv(input_csv_file)

    # Process each row in the input DataFrame
    for index, row in df.iterrows():
        app_id = row['Id']

        # Fetch the reviews for the app
        app_reviews, _ = reviews(
            app_id,
            sort=Sort.NEWEST
        )

        # Create a DataFrame for the reviews
        reviews_df = pd.DataFrame(app_reviews)

        # Add 'Id' column
        reviews_df['Id'] = app_id

        
    # Save the reviews DataFrame to a CSV file
    reviews_df.to_csv(output_csv_file, index=False, encoding='utf-8')

    print(f"Reviews retrieval completed and saved to {output_csv_file}.")

```

```python
# Call the function to fetch and save reviews
scrap_reviews('E-commerce_apps_info.csv', 'reviews.csv')
```

## Data Visualization Report

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Bar chart for Number of Installs
plt.figure(figsize=(10, 5))
sns.barplot(x='title', y='number_of_installs', data=df)
plt.xticks(rotation=45, ha='right')
plt.yticks(np.linspace(min(df['number_of_installs'] ),max(df['number_of_installs']),10), rotation=45)

plt.title('Number of Installs of Specified E-commerce App')
plt.show()
```

```python
# Histogram for Ratings
plt.figure(figsize=(8, 5))
sns.histplot(df['Average_rating'], bins=20, kde=True)
plt.title('Distribution of Ratings for E-commerce Apps')
plt.show()
```

```python
# Pie chart for Genre Distribution
plt.figure(figsize=(4, 4))
df['genre'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Genre Distribution of E-commerce Apps')
plt.show()
```

```python
# bar chart for Genre Distribution
plt.figure(figsize=(6, 4))

genre_counts = df['genre'].value_counts()
plt.bar(genre_counts.index, genre_counts.values,color=sns.palettes.mpl_palette('Dark2'))
plt.title('Genre Distribution of E-commerce Apps')
plt.xlabel('Genre')
plt.xticks(rotation=45, ha='right')
plt.show()
```
